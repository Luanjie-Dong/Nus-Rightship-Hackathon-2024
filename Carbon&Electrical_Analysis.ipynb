```python
def determine_operating_mode(row):
    if (row['speed']) > 1 or (row['nav_stat'] in [0, 3, 4, 8, 11, 12]):
        if row['maneuvering_zone'] != "PPT Maneouvering Zone" and row['in_portboundary'] == True:
            return 'Transit'
        elif row['maneuvering_zone'] == "PPT Maneouvering Zone" :
            return 'Maneuvering'
    elif row['anchorage'] == "PPT Anchorage" and (row['speed'] < 1):
        return 'Anchorage'
    elif row['berth'] in ['PPT Berth 8', 'PPT Berth 9', 'PPT Berth 1', 'PPT Berth 3',
       'PPT Berth 10', 'PPT Berth 4', 'PPT Berth 2', 'PPT Berth 7',
       'PPT Berth 5', 'PPT Berth 6'] and row['speed'] < 1:
        return 'Alongside/Hotel'
    else:
        return None  # You can change this to a default value if needed

# Apply the function to create the 'operating_mode' column
df['operating_mode'] = df.apply(determine_operating_mode, axis=1)
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
df.sort_values(by=['imo', 'timestamp'], inplace=True)
# Initialize a new column 'time_taken' with default value None
df['time_taken'] = None

# Iterate through unique IMO numbers
for imo1 in df['imo'].unique():
    # Select rows for the current IMO
    imo_rows = df[df['imo'] == imo1]

    # Calculate timedelta for each row
    imo_rows['time_taken'] = imo_rows['timestamp'].diff(periods=-1)
    imo_rows['time_taken'] = imo_rows['time_taken'].dt.total_seconds().apply(lambda x: min(-x, 10000))

    # Update the 'time_taken' column in the original DataFrame
    df.loc[df['imo'] == imo1, 'time_taken'] = imo_rows['time_taken']
    
```

# Total Electrical Load Analysis

```python
df[df.operating_mode == "Alongside/Hotel"]["timestamp_24h"]
```

```python
# Total electrical demand from ship in alongside mode (Tael) 
# = time-weighted average electrical load (ael) in alongside mode 
# = (Sum of Product of Ael multiplied by alongside duration hours) 
# divide by (Sum of total alongside duration in hours).

# Alongside or hotel load is the electrical load when the vessel is at the terminal/berth.

def electric(row):
    if row["operating_mode"] == 'Alongside/Hotel':
        total = row["ael"] * (row["time_taken"]/3600)
        return total

    
df["electric"] = df.apply(electric,axis=1)
        
```

```python
# df['electric'].sum()
# df["time_taken"].sum()

total_demand = df['electric'].sum()/(df["time_taken"].sum()/3600)
print(total_demand)
```

# Load demand variation from the grid through day and night on a monthly basis

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])

# Format 'timestamp' to 24-hour format
df['timestamp_24h'] = df['timestamp'].dt.strftime('%H%M%S').astype(int)
```

```python
df.info()
```

```python
def electric_day(row):
    if row["operating_mode"] == 'Alongside/Hotel':
        if 90000 <= row["timestamp_24h"] <= 170000 :
            total = row["ael"] * (row["time_taken"]/3600)
            return total

    
df["electric_day"] = df.apply(electric_day,axis=1)
```

```python
def electric_night(row):
    if row["operating_mode"] == 'Alongside/Hotel':
        if row["timestamp_24h"] > 170000 or row["timestamp_24h"] < 90000:
            total = row["ael"] * (row["time_taken"]/3600)
            return total

    
df["electric_night"] = df.apply(electric_night,axis=1)
```

```python
def electric_night_time(row):
    if row["operating_mode"] == 'Alongside/Hotel':
        if row["timestamp_24h"] > 170000 or row["timestamp_24h"] < 90000:
            return (row["time_taken"]/3600)

    
df["electric_night_time"] = df.apply(electric_night_time,axis=1)
```

```python
def electric_day_time(row):
    if row["operating_mode"] == 'Alongside/Hotel':
        if 90000 <= row["timestamp_24h"] <= 170000 :
            return (row["time_taken"]/3600)

    
df["electric_day_time"] = df.apply(electric_day_time,axis=1)
```
